<p class=MsoNormal align=center style='text-align:center'><b><span lang=HR
style='font-size:18.0pt;line-height:107%;font-family:"Times New Roman",serif'>DISKRETNA FOURIEROVA TRANSFORMACIJA (DFT)</span></b></p>

<p class=MsoNormal><span lang=HR style='font-family:"Times New Roman",serif'>&nbsp;</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Bilo koji <b>aperiodski</b> vremenski diskretni niz $x[n]$ s uzorcima od $−\infty$ do $\infty$ do može se
prikazati u frekvencijskoj domeni pomoću para izraza:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'>
<table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$X(e^{j\omega}) = \sum_{n=-\infty}^\infty x[n]e^{-j\omega n} 
$$</td><td bgcolor="#FFFFFF">(2.1)</td></tr>
    <tr ><td bgcolor="#FFFFFF">$$x[n] =\frac{ 1}{2\pi}\int_{-\pi}^{\pi} X(e^{j\omega n})e^{j\omega n}d\omega
$$</td><td bgcolor="#FFFFFF">(2.2)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>U gornjim izrazima $X(e^{j\omega n})$ predstavlja Fourierovu transformaciju diskretnih signala (<i>engl. Discrete Time Fourier Transform</i>, DTFT, a koristi se i kraći termin spektar). Fourierova transformacija postoji ako suma u izrazu (2.1) uniformno konvergira, a to uvijek vrijedi ako je suma apsolutnih vrijednosti niza $x[n]$ za sve indekse $n$konačna, tj. </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'>
<table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$ \sum_{n=-\infty}^\infty |x[n]| < \infty $$</td><td bgcolor="#FFFFFF">(2.3)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Dobivena transformacija $X(e^{j\omega n})$ je tada kontinuirana funkcija od $\omega$ i periodična s $2\pi$.</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Za bilo koji beskonačni <b>periodični</b> vremenski niz gornji uvjet (2.1) nije zadovoljen i Fourier-ova transformacija u striktnom smislu ne postoji kao kontinuirana funkcija već sadrži sumu delta funkcija na frekvencijama svih harmonika tog periodičnog signala (vidi kasnije). Stoga se takvi signali $\tilde{x}[n]$ ($\tilde{}$ naglašava periodičnost) perioda $P$ često prikazuju pomoću sume $P$ diskretnih kompleksnih eksponencijala frekvencija:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
 \omega_k = \frac{2\pi k}{P}\text{, za }k=0,...,P-1.
$$</td><td bgcolor="#FFFFFF">(2.4)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>što je dano izrazom:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px"><tr ><td bgcolor="#FFFFFF">$$
\tilde{x}[n] = \frac{1}{P} \sum_{k=0}^{P-1}c_k e^{j\omega_k n} =\frac{1}{P} \sum_{k=0}^{P-1}c_k e^{j\frac{2\pi k n}{P} n} \text{ , } k=0,1,...,P-1.$$</td><td bgcolor="#FFFFFF">(2.5)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Koeficijente $c_k$ razvoja u red, tj. kompleksne amplitude pripadnih eksponencijala moguće je naći prema izrazu: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
c_k = \sum_{n=0}^{P-1} \tilde{x}[n] e^{j\frac{2\pi kn}{P}} \text{ , }k=0,1,...,P-1.
$$</td><td bgcolor="#FFFFFF">(2.6)</td></tr></table></span></p></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Zbog analogije sa rastavom periodičnih vremenski kontinuiranih signala, često se izraz (2.5) naziva diskretni Fourierov red (engl. Discrete Fourier Series, DFS). Kod toga osim periodičnosti diskretnog niza u vremenskoj domeni $x[n] = x[n + rP]$ ($r$ cijeli broj) vrijedi i periodičnost koeficijenata razvoja $c_k = c_{k+rP}$. Bilo kojih $P$ susjednih uzoraka, bilo vremenskog niza, bilo njegovih koeficijenata razvoja jednoznačno opisuje taj periodični niz. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Pomoću ovako određenih koeficijenata diskretnog Fourierovog reda moguće je napisati čak i izraz za Fourierovu transformaciju $X(e^{j\omega})$ takvog periodičnog signala beskonačnog trajanja i beskonačne energije i to pomoću izraza:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X(e^{j\omega} = \sum_{r=-\infty}^{\infty}\sum_{k=0}^{P-1}\frac{2\pi}{P}\cdot c_k \cdot \delta (\omega - \omega_k + 2\pi \cdot r)
$$</td><td bgcolor="#FFFFFF">(2.7)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Ovaj izraz se ne može dobiti na osnovu (2.1) zbog problema konvergencije, ali se može pokazati da se njegovim uvrštavanjem u izraz za inverznu Fourierovu transformaciju (IDTFT) (2.2) dobiva polazni beskonačni periodični signal $\tilde{x}[n]$.</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Fourierova transformacija vremenski diskretnih aperiodskih signala konačnog trajanja je kontinuirana funkcija od frekvencije $\omega$, prema (2.1). Radi predstavljanja takve funkcije na računalima pogodno je koristiti njene uzorke, a ne funkciju. Matematički postupak kojim je
moguće odrediti <b>$N$ uzoraka Fourierove transformacije</b> nekog <b>aperiodskog</b> signala konačnog trajanja naziva se diskretna Fourierova transformacija (engl. <i>Discrete Fourier Transform</i>, DFT) i dana je izrazima: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X[k] = \sum_{n=0}^{N-1}x[n]e^{-j\frac{2\pi n k}{N}}, $$ $$
x[n] = \frac{1}{N}\sum_{k=0}^{N-1}X[k]e^{-j\frac{2\pi n k}{N}}
$$</td><td bgcolor="#FFFFFF">(2.8)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Radi jednostavnosti izrazi se često zapisuju u obliku: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
x[n] = \frac{1}{N}\sum_{k=0}^{N-1}X[k]W_N^{-nk}\text{ , }X[k] = \sum_{n=0}^{N-1}x[n]W_N^{nk}
$$</td><td bgcolor="#FFFFFF">(2.9)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje je $W_N$ dan izrazom: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$W_N = e^{-j\frac{2\pi}{N}}.
$$</td><td bgcolor="#FFFFFF">(2.10)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Otipkavanje Fourier-ove transformacije izvedeno je u N ekvidistantnih točaka s frekvencijama $\omega_k$, $0 \leq \omega_k \leq 2\pi$, tj. vrijedi: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X[k] = X(e^{j\omega})\mid_{\omega = \omega_k = \frac{2\pi k}{N}} \text{, } k=0,...,N-1.
$$</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Drugim riječima, koeficijenti DFT transformacije $X[k]$ predstavljaju uzorke od $X(e^{j\omega}$ na frekvencijama $\omega_k$. Ulaz u DFT su uzorci diskretnog niza sa indeksima $n = 0,1,...,N-1$. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Treba napomenuti da broj uzoraka DFT-a mora biti jednak ili veći od dužine aperiodskog niza $R$, dakle $N \geq R$. Ako je $N$ strogo veći od $R$, niz tada treba nadopuniti nulama tako da je konačna dužina niza jednaka $N$. Za kauzalne je nizove postupak očit jer nule treba staviti iza zadnjeg uzorka originalnog niza, ali je situacija nešto složenija za nekauzalne nizove. Kod njih treba odgovarajući broj nula staviti također iza zadnjeg uzoraka s pozitivnim indeksima, ali nakon nula treba staviti uzorke s negativnim indeksima i to tako da uzorak sa indeksa $-1$ dolazi na indeks $N-1$, sa indeksa $-2$ na $N-2$ i tako dalje. Na taj način je jedan <b>nekauzalni aperiodski niz pripremljen za analizu primjenom DFT-a</b> uz očuvanje njegovih faznih svojstava. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Uz uvjet $N > R$ se DFT-om dobije <b>više uzoraka očitanog spektra</b> što daje vjerniju sliku o samom spektru niza $X(e^{j\omega}$. Da bi i uz veliki broj uzoraka DFT-a izračun bio brz, razvijen je niz algoritama koji su po krajnjem rezultatu potpuno ekvivalentni DFT-u, ali su učinkovitiji, a time i brži. Takvi algoritmi nazivaju se jednim imenom brza Fourierova transformacija (engl. <i>Fast Fourier Transform</i>, FFT). </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>S obzirom da izrazi za DFS (2.5) i (2.6) definirani za periodične diskretne signale izgledaju identično kao oni za DFT (2.8) i (2.9) kojim se predstavljaju aperiodski diskretni signali konačnog trajanja (osim imena $c_k$ i $X[k]$, te periode $P$ odnosno broja uzoraka $N$), postavlja se pitanje njihove veze. Može se pokazati da se svaki aperiodski diskretni signal trajanja $N$ može pretvoriti u beskonačno periodičan tako da se ponavlja dotičnih $N$ uzoraka niza, a sam aperiodski signal se uvijek može dobiti kao samo jedna perioda takvog periodiziranog signala. Takav postupak periodizacije u vremenskoj domeni rezultira diskretnim spektrom pa samim time i razvoj u DFS ima smisla. Možemo zaključiti da u definiciji DFT reprezentacije aperiodskog signala jednostavno prepoznajemo da nas interesiraju samo vrijednosti $x[n]$ u intervalu$ n = 0, 1, N −1 $jer znamo da su druge jednake nuli. Isto tako nas interesiraju samo vrijednosti $X[k]$ u intervalu $k = 0, 1, N −1$ jer su one jedine potrebne u izrazu (2.9). Međutim, ono što je važno i treba zapamtiti jest da iako je DFT <i>de facto</i> transformacija koja se koristi u praksi, rastav u DFS predstavlja teoretsku podlogu. <b>Inherentna periodičnost DFS transformacije ostaje i kod DFT-a, i u vremenskoj i u frekvencijskoj domeni</b>. To drugim riječima znači da iako znamo da imamo aperiodski signal, inverznim DFT-om on biva periodiziran. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Iako se po definiciji koristi za konačne aperiodske nizove, <b>pomoću DFT-a možemo odrediti i harmonijski sastav beskonačnog periodičnog signala</b> (tj. koeficijente njegovog DFS rastava) pod uvjetom da je <b>broj uzoraka DFT-a $N$ jednak ili cjelobrojni višekratnik periode $P$</b> tog periodičnog niza. Ako to nije zadovoljeno, uzorci DFT-a $X[k]$ ne odgovaraju koeficijentima $c_k$ i ne pokazuju očekivani harmonijski sastav. Jednostavno objašnjenje toga je činjenica da DFT pretpostavlja periodičko ponavljanje upravo tih N uzoraka, a pošto nije obuhvaćen cijeli broj perioda ulaznog niza, periodizirani niz ima neočekivane skokove na rubovima ponovljenih blokova. Činjenica da se radi o potpuno drugom periodičnom signalu od onog kojeg smo imali na početku uzrokuje i pojavu drugačijeg harmonijski sastava. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Za uzorke DFT-a moguće je izračunati amplitudu $|X[k]|$ i fazu $\angle X[k]$ koji predstavljaju otipkanu amplitudnu i faznu karakteristiku (spektar) od $X(e^{j\omega}$ prema izrazima:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
|X[k]| = \sqrt{Re{X[k]}^2+Im{X[k]}^2} \text{, } \angle X[k] = atan2(Im{X[k]},Re{X[k]})
$$</td><td bgcolor="#FFFFFF">(2.11)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje je sa atan2 označena 4-kvadrantna arctg funkcija od imaginarnog i realnog dijela kompleksnih uzoraka DFT-a $X[k]$.</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>U nastavku će biti opisana neka od svojstava DFT-a. </span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>I. Realni vremenski niz</h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Koeficijenti DFT-a realnog vremenskog niza imaju slijedeće svojstvo </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X[k] = X\ast [-k]
$$</td><td bgcolor="#FFFFFF">(2.12)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>tj. koeficijenti, simetrični oko 0, su konjugirano kompleksni, odnosno realni dio im je jednak, a imaginarni dio suprotnog predznaka</span></p>

<<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px"
    <tr ><td bgcolor="#FFFFFF">$$
Re{X[k]} = Re{X[-k]} \text{, } Im{X[k]} = -Im{X[-k]}.
$$</td><td bgcolor="#FFFFFF">(2.13)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Zapisano preko modula (magnitude) i faze </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px"
    <tr ><td bgcolor="#FFFFFF">$$
|X[k]| = |X[-k]| \text{, } \angle X[k]=-\angle X[-k]
$$</td><td bgcolor="#FFFFFF">(2.14)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>proizlazi da simetrični koeficijenti imaju jednake module, a pripadne faze su suprotnog predznaka. Zbog toga je za realne vremenske nizove dovoljno poznavati samo prvih $\frac{N}{2} + 1$ koeficijenata DFT-a. </span></p>

<h4> Napomena </h4>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Važno je ovdje prokomentirati što u stvari znači simetričnost ili antisimetričnost DFT-a. Već je rečeno da DFT za ulaz ima $N$ uzoraka vremenskog niza između $0$ i $N-1$. Rezultat DFT je niz uzoraka $X[k]$ koji su također između $0$ i $N-1$. Prema izrazu (2.12) ti uzorci odgovaraju frekvencijama $0 \leq \omega_k \leq 2\pi$. Simetričnost/antisimetričnost je klasično definirana s obzirom na ishodište (u ovom slučaju, frekvenciju 0) pa bi promatrani raspon frekvencija trebao u stvari biti $-\pi \leq \omega_k \leq \pi$. Vidimo da uzorci DFT ne pokrivaju taj raspon. Ovdje stoga treba uzeti u obzir periodičnost DFT-a što znači da su uzorci DFT-a koji odgovaraju frekvencijama $\pi \leq \omega_k \leq 2\pi$ upravo jednaki onima u rasponu $-\pi \leq \omega_k \leq 0$. Dovoljno je, dakle, zamisliti da se <b>desna polovica DFT uzoraka na indeksima $k$ od $N/2$ do $N-1$ (za $N$N paran) nalazi i ijevo od ishodišta, na indeksima $-N/2$ do $-1$</b>. Uz ovo gornji uvjeti imaju smisla. </span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>II. Simetrični/antisimetrični vremenski nizovi</h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Ako je realni vremenski niz dužine $N$ <b>simetričan</b> (paran), tj. vrijedi: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
x[n] = x[-n] \text{, za} n=1,...,\frac{N}{2}-1
$$</td><td bgcolor="#FFFFFF">(2.15)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>tada su koeficijenti DFT-a <b>čisto realni</b> (imaginarni dio je jednak 0, $Im{X[k]}=0$).</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>
Analogno, ako je vremenski niz <b>antisimetričan</b> (neparan):</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
x[n] =-x[-n] \text{, za} n=1,...,\frac{N}{2}-1 \text{ i } x[0]=0
$$</td><td bgcolor="#FFFFFF">(2.16)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>tada su koeficijenti DFT-a čisto imaginarni (realni dio je jednak 0, $Re{X[k]}=0$).</span></p>

<h4>Napomena</h4>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Ponovno treba dodati komentar da bi uvjet bio jasniji. Definicija simetričnosti općenito je vezana uz izgled signala s lijeve i desne strane ishodišta. Nasuprot tome, ulazni niz u DFT mora biti definiran za indekse od $0$ do $N-1$. Ovdje također možemo primijeniti inherentnu periodičnost DFT transformacije i to onaj dio koji se odnosi na periodičnost u vremenskoj domeni. Naime, $N$ uzoraka ulaznog niza predstavlja samo ono što gledamo, međutim DFT podrazumijeva da se tih $N$ uzoraka periodički ponavlja u $±$ beskonačnost. Ako se provede takva periodizacija, dobivamo signal koji s obzirom na ishodište može biti simetričan, asimetričan ili niti jedno od toga. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Iz ove diskusije jasniji je i način kako ostvariti <b>DFT nekauzalnog aperiodskog niza</b> koji ima uzorke i lijevo i desno od nultog indeksa. Kao što je rečeno, da bismo uopće mogli izračunati DFT tog niza, potrebno ga je prikazati između uzoraka $0$ do $N-1$. Ako nekauzalni niz periodiziramo, uzorci sa negativne osi pojavit će se oko indeksa $N-1$ i imat ćemo niz od kojeg možemo raditi DFT. </span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>III. Pomak vremenskog niza</h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Najprije razmotrimo svojstvo pomaka vremenskog niza za Fourier-ovu transformaciju vremenski diskretnog signala (<b>DTFT</b>). Uzmimo diskretni niz $x[n]$, te ostvarimo drugi niz, pomaknut u odnosu na ovaj originalni za $m$ koraka ($m$ je cijeli broj), $x_m[n]=x[n-m]$. Ako je DTFT prvog niza jednak $X(e^{j\omega})$, tada za DTFT niza pomaknutog za $m$ koraka $X_m(e^{j\omega})$, vrijedi: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
X_m(e^{j\omega})=X(e^{j\omega})e^{-k\omega m}.
$$</td><td bgcolor="#FFFFFF">(2.17)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Pošto je modul od $e^{-j\omega m}$ jednak $1$, moduli DTFT-a originalne i pomaknute sekvence su jednaki, a faze se razlikuju, jer član $e^{-j\omega m}$ unosi linearni fazni pomak. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Za <b>DFS transformaciju</b> koja podrazumijeva ulazni beskonačno <b>periodični</b> niz $\tilde{x}[n]$ perioda $P$ vrijedi analogno svojstvo koje se može opisati izrazom: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF" width="200px">$$
c_{k,m}=c_k \cdot e^{-j\frac{2\pi k}{P}m}=c_k \cdot W_P^{k \cdot m}
$$</td><td bgcolor="#FFFFFF">(2.18)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje su $c_k$ i $c_{k,m}$, DFS koeficijenti originalnog niza $\tilde{x}[n]$, te niza pomaknutog za $m$ koraka $\tilde{x}[n-m]$. Sam pomak niza ne predstavlja nikakav problem s obzirom da je niz periodičan pa znamo da uzorci postoje i izvan raspona od $0$ do $P-1$.  </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'><b>DFT transformacija</b> je definirana za aperiodske diskretne nizove $x[n]$ dužine $N$. Za nju također vrijedi gornje svojstvo koje se može zapisati kao: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
X_m[k] = X[k]W_N^{km}
$$</td><td bgcolor="#FFFFFF">(2.19)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje su $X[k]$ i $X_m[k]$ DFT transformacije originalnog $x[n]$ niza s vrijednostima definiranim za indekse $n = 0,..., N −1$, te niza pomaknutog za $m$ koraka $x[n-m]$. U slučaju DFT-a, međutim, treba biti jako oprezan što znači pomak niza. Pošto se radi o aperiodskom signalu, definiranom od 0 do $N-1$, a za sve druge vrijednosti jednak 0, to bi možda značilo da je pomakom u vremenu dio originalnog niza zauvijek izgubljen. To naravno nije tako posebice ako se podsjetimo na periodičnost u vremenu 'naslijeđenu' od DFS transformacije, a koja vrijedi i za DFT. Naime, ako originalni aperiodski niz u glavi periodiziramo s periodom $N$, pomaknemo za $m$ koraka i od takvog pomaknutog, periodiziranog niza uzmemo samo uzorke na indeksima od 0 do $N-1$, dobili smo pomaknuti aperiodski niz. Ono što se de facto događa jest da se <b>aperiodski niz pomiče cirkularno unutar $N$ uzoraka</b> što odgovara operaciji modulo $N$. Dakle svojstvo (2.21) vrijedi za signal $x_m[n]=x[(n-m)_{mod N}].$</span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>IV. Signali u obliku periodičnih impulsa</h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Kod periodičnih signala koji imaju oblik niza impulsa, tj. kod kojih je mnogo uzoraka jednakih 0, spektar je širok. Gotovo svi uzorci DFS-a su različiti od 0. Vrijedi i dualno svojstvo, tj. da što je impuls u vremenu širi, to mu je spektar uži (sve više koeficijenata DFS-a je približno jednako 0). Diskretna periodična funkcija pogodna za promatranje ove pojave ima oblik Gauss-ove funkcije i definirana je na jednom periodu $P$ izrazom:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
x[n] =\frac{1}{\sigma \sqrt{2\pi}}e^{-\frac{2(n-P/2)^2}{P^2\sigma^2}} \text{, } n=0,...,P-1.
$$</td><td bgcolor="#FFFFFF">(2.20)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Promjenom parametra $\sigma$ mijenja se širina Gauss-ovog zvona (što je $\sigma$ manji, zvono je uže). Modul koeficijenata DFS-a ima ponovno oblik zvona, ali što je zvono u vremenu uže, zvono u Fourierovoj domeni je šire i obratno. </span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>V. DFT diskretnog kosinusnog niza </h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Promatrajmo vremenski diskretni kosinusni niz amplitude $A$, početne faze $\phi_0$ i frekvencije $\omega_0$, danog sa: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
\tilde{x}[n] = Acos(\omega_0 n+\phi_0)=Acos(2\pi/P \cdot m_0 \cdot n + \phi_0) \text{, za} n = -\infty \text{ do } \infty
$$</td><td bgcolor="#FFFFFF">(2.21)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje su $m_0$ i $P$ cijeli brojevi, $0  \leq m_0 \leq P/2$. Broj $P$ predstavlja osnovni period takvog kosinusnog niza ili njegov višekratnik u slučaju da $m_0$ i $P$ nisu koprim brojevi. Takav signal
sadrži točno $m_0$ cijelih perioda kosinusnog signala unutar $P$ susjednih uzoraka. Jasno je da se radi o beskonačnom periodičnom nizu. U skladu sa već rečenim (vidi izraz (2.7)), Fourierova transformacija (spektar, DTFT) $X(e^{j\omega})$ takvog niza ima oblik:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
X(e^{j\omega})=\sum_{r=-\infty}^{\infty}2\pi \cdot (\frac{A}{2}e^{j\phi_0} \cdot \delta (\omega - \omega_0 + 2\pi \cdot r) + \frac{A}{2}e^{-j\phi_0} \cdot \delta (\omega + \omega_0 + 2\pi \cdot r),
$$</td><td bgcolor="#FFFFFF">(2.22)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>tj. ima delta funkcije na frekvencijama $\omega = ±\omega_0 + 2r\pi, -\infty \leq r \leq \infty$, čije su kompleksne amplitude određene amplitudom $A$ i početnom fazom $\phi_0$, dok je za sve ostale frekvencije $X(e^{j\omega})$ jednako 0.</span></p> 

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Da bismo uopće mogli izračunati DFT, moramo ograničiti kosinusni niz i promatrati samo jedan njegov dio. To ograničavanje niza u vremenskoj domeni odgovara množenju kosinusnog niza sa uzorcima jednog interesantnog i korisnog tipa vremenski diskretnog niza koji se naziva <b>vremenski otvor</b> (engl. <i>window</i>). Postoji više vrsta vremenskih otvora koji se mogu koristiti za ovu namjenu, a najjednostavniji je pravokutni vremenski otvor (engl. rectangular window). Pravokutni otvor definiran je sa:</span></p> 

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
w[n] = \begin{cases}
1 & \text{ }n=0,...,L-1 \\ 0 & \text{ inače}
\end{cases}
$$</td><td bgcolor="#FFFFFF">(2.23)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>tj. on ima uzorke jednake 1 samo za L uzoraka dok su svi ostali jednaki 0. Množenjem beskonačnog, kosinusnog niza sa ovim otvorom $w[n]$ dužine $L$ uzoraka dobiva se konačan, aperiodski diskretni niz $x_L[n]$ koji ima vrijednost 0 za $n < 0$ i $n \geq L$ , a za $n=0,1,...,L-1$ ima uzorke čije su vrijednosti jednake onima originalne kosinusoide $\tilde{x}[n]$. U praksi, ako se za ograničavanje signala koristi pravokutni vremenski otvor nije potrebno provoditi množenje već se jednostavno izdvoji L uzoraka te kosinusoide. </span></p> 

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>U spektralnoj domeni, operacija množenja uzoraka niza i vremenskog otvora odgovara tzv. periodičkoj konvoluciji danoj izrazom: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
X_L(e^{j\omega})=\frac{1}{2\pi}\int_{-\pi}^{\pi}X(e^{j\Phi}) \cdot W(e^{j(\omega - \Phi)})d\Phi
$$</td><td bgcolor="#FFFFFF">(2.24)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>gdje je $X_L(e^{j\omega})$ spektar ograničenog niza, $X(e^{j\omega})$ spektar periodičnog niza dan izrazom (2.24), a $W(e^{j\omega})$ spektar vremenskog otvora. Za pravokutni vremenski otvor vrijedi izraz:</span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="400px">
    <tr ><td bgcolor="#FFFFFF">$$
W(e^{j\omega})=\sum_{n=0}^{L-1}1\cdot e^{-j\omega n}=e^{-j\omega \cdot (L-1)/2} \cdot \frac{sin(\omega L/2)}{sin(\omega /2)}
$$</td><td bgcolor="#FFFFFF">(2.25)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>$W(e^{j\omega})$ je krivulja s glavnom laticom i bočnim laticama (titrajima) koja prolazi kroz 0 u točkama s frekvencijom koje su višekratnici frekvencije $2\pi /L$ , dok je vrijednost spektra otvora za $\omega = 0$ jednaka $L$. Gornji izraz opisuje spektar kauzalnog pravokutnog otvora pozicioniranog na indeksima 0 do $L-1$. Eksponencijalni član ispred omjera sinusnih funkcija posljedica je pomaka otvora u vremenu za $(L-1)/2$ koraka u desno i predstavlja linearnu faznu karakteristiku s nagibom $-(L-1)/2$. Periodičkom konvolucijom se periodični <b>spektar otvora ponavlja na svim frekvencijama na kojima se nalaze delta funkcije</b> od $X(e^{j\omega})$ u rasponu frekvencija $\omega ∈ [-\pi,\pi)$ te tako nastaje spektar $X_L(e^{j\omega})$ dan izrazom: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X_L(e^{j\omega})=\frac{A}{2}e^{j\phi_0} \cdot W(e^{j(\omega - \omega_0)})+\frac{A}{2}e^{j\phi_0} \cdot W(e^{j(\omega + \omega_0)}).
$$</td><td bgcolor="#FFFFFF">(2.26)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Ako od $L$ uzoraka $x_L[n]$, $n=0,...,L-1$ sada napravimo DFT u $N=L$ točaka, dobivamo $N$ kompleksnih koeficijenata $x_L[k]$ , $k=0,...,N-1$ koji su identični vrijednostima funkcije $X_L(e^{j\omega})$ za $\omega = 2\pi k/N$. Varijabla k se u kontekstu DFT-a često u engl. literaturi naziva bin. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>U slučaju kada vrijedi <b>$L = P = N$</b>, tj. kada smo vremenskim otvorom izdvojili točno jedan osnovni period signala koji sadržava $m_0$ perioda diskretne kosinusoide, te kada od tih $L$ uzoraka izračunamo DFT u istom broju točaka, uzorci DFT-a su jednaki: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
X[k] = 
\begin{cases}L \cdot A/2 \cdot e^{j\phi_0}& \text{  }k=m_0 \\
                L \cdot A/2 \cdot e^{-j\phi_0}& \text{  }k=N-m_0 \\
                0 &\text{   ostale }k 
                \end{cases}
                $$</td><td bgcolor="#FFFFFF">(2.27)</td></tr></table></span></p>
                
<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>Samo uz ovaj odnos $L$,$P$ i $N$, <b>uzorci DFT-a upravo odgovaraju koeficijentima DFS razvoja</b> pa izraz (2.29) predstavlja osnovu za korištenje DFT-a u svrhu analize spektralnog sastava periodičnih signala jer je vidljivo da je u koeficijentima DFT-a sadržana informacija i o amplitudi i o početnoj fazi svakog pojedinog harmonika ulaznog niza. U slučaju da se diskretni niz sastoji od više harmonijskih komponenata zajedničkog osnovnog perioda $P$, tada
svakoj kosinusoidi na binu mr odgovara jedan par DFT koeficijenata $X[m_r]$ i $X[N-m_r]$. Gornji izraz ne vrijedi u slučaju istosmjernog signala ($m_0=0$) ili signala frekvencije $\omega_0 = \pi$ tj., $m_0 = P/2$. Tada se dvije kompleksne eksponencijale preklapaju pa postoji samo jedan uzorak DFT-a čija je amplituda jednaka $LA$.  </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>U slijedeća dva paragrafa bit će analiziran <b>utjecaj broja uzoraka DFT-a $N$</b> na interpretaciju samih DFT uzoraka. Ako sa $L$ uzoraka obuhvatimo točno jedan osnovni period niza $\tilde{x}[n]$ (tj. vrijedi $P=L$), a izračunajmo <b>DFT u broju točaka $N$, koji je višekratnik od $L$</b>, tj. $N=r \cdot L$ (nadopunjavanjem niza s nulama), tada će svaki $r$-ti uzorak novog DFT-a biti jednak uzorcima originalnog DFT-a izračunatog u $L$ točaka. Između tih $r$-tih višekratnika pojavit će se $r-1$ novih uzoraka koji su različiti od 0. Ako je npr. $N=2L$, između parnih uzoraka, pojavit će se još po jedan novi različit od 0. Ova situacija de facto odgovara <b>gušćem otipkavanju spektra</b> aperiodskog signala $x_L[n]$, čime se izgledom skupa uzoraka DFT-a približavamo stvarnom izgledu samog spektra $X_L(e^{j\omega})$. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Ako je uz isti uvjet $P=L$, $N$ <b>proizvoljni broj veći od $L$</b> (dakle nije njegov višekratnik), i dalje se DFT-om dobivaju uzorci spektra $X_L(e^{j\omega})$ u ekvidistantnim točkama $\omega_k = 2\pi /N\cdot k$, $k=0,...,N-1$, ali se ove frekvencije više ne poklapaju s frekvencijama harmonika ulaznog signala. Krivulja $X_L(e^{j\omega})$ je dakle otipkana na mjestima koja nam nisu pogodna za određivanje spektralnog sastava signala. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Na kraju, analizirajmo i utjecaj <b>veličine vremenskog otvora $L$</b> na značenje dobivenih DFT uzoraka. Ako je sa $L$ uzoraka ograničenog kosinusnog niza $x_L[n]$ obuhvaćeno <b>više cijelih osnovnih perioda $P$</b> originalnog periodičnog niza od kojeg se krenulo, tj. vrijedi $L=i \cdot P$ gdje je $i$ prirodni broj, kružna frekvencija $\omega_0$ ulaznog diskretnog niza $\tilde{x}[n]$ se može
izraziti kao: </span></p>

<p class=MsoNormal align=center style='margin-bottom:0in;margin-bottom:.0001pt;
text-align:center;text-indent:.5in;line-height:normal'><span lang=HR
style='font-size:12.0pt;font-family:"Times New Roman",serif'><table width="595px">
    <tr ><td bgcolor="#FFFFFF">$$
\omega_0=\frac{2\pi}{P}\cdot m_0=\frac{2\pi}{L}\cdot i\cdot m_0=\frac{2\pi}{L}\cdot l_0 \text{ , }l_0\text{ je cijeli broj,}
$$</td><td bgcolor="#FFFFFF">(2.28)</td></tr></table></span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;line-height:normal'><span lang=HR style='font-size:12.0pt;font-family:
"Times New Roman",serif'>dakle jednaka je i višekratniku od $2\pi/L$. Taj višekratnik $l_0$ predstavlja upravo broj cijelih perioda diskretne kosinusoide na $L$ uzoraka. U slučaju da se DFT izvodi u $N$ točaka uz $N=L$, uzorci amplitude DFT-a dotičnog niza imati će sve uzorke jednake nuli osim onih na indeksu $k=0l_0$ i $N-l_0$. Drugim riječima, za neki periodični niz sa proizvoljnim harmoničkim sastavom osnovnog perioda $P$, tek <b>svaki $i$-ti uzorak DFT-a odgovara DFS koeficijentima</b>. </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Razmotrimo i slučaj kada uzorcima $x_L[n]$ od 0 do $L −1$ <b>nemamo obuhvaćen cijeli broj perioda</b> od $\tilde{x}[n]$ , tj. odnos $L$ i $P$ je proizvoljan. Tada kružna frekvencija $\omega_0$ diskretnog
niza $\tilde{x}[n]$ više nije višekratnik od $2\pi/L$. Ako je ulazni niz bio samo jedna kompleksna eksponencijala zadane frekvencije $\omega_0$, to ne izaziva neke poteškoće. Problem je međutim ako
imamo proizvoljni periodični niz osnovnog perioda $P$ koji se sastoji od $P$ takvih kompleksnih eksponencijala na frekvencijama $\omega_0 = 2\pi/P\cdot m_0$, $m_0=0,...,P-1$. Diskretna kosinusoida koja se analizira u ovom tekstu, također spada u ovu klasu jer se sastoji od dvije takve kompleksne eksponencijale (prema $cos(\omega_0 n)=1/2\cdot (e^{j\omega_0 n)} + e^{-j\omega_0 n})$ jedna eksponencijala je na pozitivnoj frekvenciji, a druga na negativnoj frekvenciji). Uslijed množenja sa vremenskim otvorom proizvoljne širine $L$ u odnosu na $P$, spektar otvora ne prolazi kroz 0 na frekvencijama harmonika pa dolazi do međusobne interakcije harmonika u spektru $X_L(e^{j\omega})$. Ova pojava onemogućava analizu harmonijskog sastava takvog periodičnog signala primjenom DFT-a. Pojava se naziva <b>'razmazivanje'</b> spektra, a može se interpretirati i činjenicom da se periodizacijom aperiodskog niza $x_L[n]$ u stvari dobiva neki novi periodični
signal, a ne polazni periodični niz $\tilde{x}_L[n]$.</span></p>

<h3 class=MsoNormal style='text-autospace:none'><b><span lang=HR style='font-size:18.0pt;line-height:106%;font-family:"Times New Roman",serif'>Realizacija na DSP maketi</h3>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>U dijelu vježbe s DSP maketom pokazat će se usporedba idealnih frekvencijski ograničenih signala i realno otipkanih. Time će se pokazati neki efekti očitavanja na izgled DFT-a. U svim promatranim slučajevima, DFT se računa na vremenski ograničenim kosinusnim signalima pa vrijedi sve rečeno pod stavkom V. iz teoretskog dijela vježbe. Kako je u vježbi broj uzoraka otvora fiksan, u općem slučaju nema cijeli broj perioda na promatranom odsječku signala pa dolazi do razmazivanja spektra (frekvencije komponenata od kojih je sastavljen signal ne podudaraju se s frekvencijama spektralnih komponenti koje postoje u DFT transformaciji). </span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:
12.0pt;font-family:"Times New Roman",serif'>Na vježbama će se promatrati dva signala istog oblika i frekvencije, ali dobivena na dva različita načina. <b>Prvi signal dobiven je očitavanjem analognog signala</b> iz generatora funkcija ugrađenog u DSP maketu. Pošto je to realan signal on osim potrebnih harmonika osnovne frekvencije, koji su sastavni dio promatranog signala, sadrži i neke druge frekvencijske komponente koje su prigušene za neki iznos u odnosu na osnovne, ali doprinose iznosu ukupnog signala (u biti uzrokuju izobličenje signala). Pojava takovih komponenata posljedica je različitih nelinearnih efekata unutar generatora funkcija. Nadalje, kao posljedica filtracije određene ulaznom karakteristikom A/D pretvornika (antialiasing filtar) dolazi do odbacivanja harmonijskih komponenti koje se nalaze iznad polovice frekvencije otipkavanja. To odbacivanje nije idealno jer radi realnog filtra ipak dolazi do preslikavanja nekih komponenti u blizini polovine frekvencije otipkavanja. Te komponente su dosta prigušene što filtrom, što samim smanjenjem iznosa harmonika te ne utječu u nekoj većoj mjeri na signal. Proučavanjem spektra otipkanog signala možemo ustanoviti kvalitetu  generatora funkcija te korištenog A/D pretvornika. Sama usporedba signala se na vježbama provodi na različitim frekvencijama za tri valna oblika (sinus, trokut i pravokutnik).</span></p>

<p class=MsoNormal style='margin-bottom:0in;margin-bottom:.0001pt;text-align:
justify;text-indent:.5in;line-height:normal'><span lang=HR style='font-size:

Matematički izrazi za promatrane valne oblike su slijedeći: </span></p>

$$
\begin{equation}\label{eq:1} \text{Sinus:  }y(t) = sin(2 \cdot \pi \cdot f \cdot t) \end{equation}\\
\text{Trokut: }y(t)=\frac{8}{\pi^2}\cdot \sum_{n=1}^{\infty}(-1)^n\frac{sin(2 \cdot \pi \cdot f\cdot (2\cdot n -1) \cdot t)}{(2\cdot n-1)^2} \\
\text{Pravokutnik: }y(t)=\frac{4}{\pi}\cdot \sum_{n=1}^{\infty}(-1)^n\frac{sin(2 \cdot \pi \cdot f\cdot (2\cdot n -1) \cdot t)}{2\cdot n-1}
$$
